```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Leitura de uma imagem colorida
f = cv2.imread('image.jpg')

# Exibio da imagem original
plt.imshow(cv2.cvtColor(f, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Converso RGB para Grayscale
R = f[:,:,0]  # matriz correspondente  componente RED
G = f[:,:,1]  # matriz correspondente  componente GREEN
B = f[:,:,2]  # matriz correspondente  componente BLUE

f2 = 0.299 * R + 0.587 * G + 0.114 * B

# Exibio da imagem em escala de cinza
plt.imshow(f2, cmap='gray')
plt.show()
```
